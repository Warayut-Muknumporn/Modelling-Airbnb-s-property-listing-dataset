```python
def remove_rows_with_missing_ratings(data):
    # Convert data to a pandas DataFrame
    #df = pd.DataFrame(data)
    
    # Get a list of columns that contain 'rating'
    rating_cols = [col for col in data.columns if 'rating' in col]
    
    # Drop rows with missing values in the rating columns
    data.dropna(subset=rating_cols, inplace=True)
    
    # Return the updated DataFrame
    return data


# Drop NaN values in these columns : 'Cleanliness_rating','Accuracy_rating','Communication_rating','Location_rating','Check-in_rating','Value_rating'

# df.dropna(subset = ['Cleanliness_rating','Accuracy_rating','Communication_rating','Location_rating','Check-in_rating','Value_rating'], inplace= True)
```

```python
def combine_description_strings(data):
    # Remove rows with missing values in the 'Description' column
    data.dropna(subset=['Description'], inplace=True)
# The r prefix in front of a string literal in Python denotes a raw string. 
# In a raw string, backslashes (\) are treated literally, meaning that they are not used as escape characters. 
# For example, the string '\n' is a two-character string consisting of a backslash and the letter 'n',
#  whereas the raw string r'\n' is a two-character string consisting of a backslash and the letter 'n'.
# In the specific case of rep_chars = r'\*|\]|\[', 
# the raw string is used to define a regular expression pattern that matches three specific characters: the asterisk (*), 
# the right square bracket (]), and the left square bracket ([).
# The | symbol in the pattern means "or", so the pattern matches any string that contains any of these three characters. 
    
    # Replace certain characters with empty strings

    rep_char = r'\*|\]|\['
    data['Description'] = data['Description'].str.replace(rep_char, '')
    
        # Replace newline and single quote characters with spaces
    data['Description'].replace([r'\\n', '\n', r"\'"], [' ', ' ', ''], regex=True, inplace=True)
    
    # Remove the phrase 'About this space' and any trailing commas or spaces
    data['Description']=df['Description'].str.strip('About this space, ')
    
    # Return the updated DataFrame
    return data


# Remove the first 'About this space' segment of the Descriptions column, and removing any apostrophes that may confusion down the line.

# df=df[dfF['Description'].str.contains('About this space')== True]
# df['Description'] = df['Description'].str.replace("'About this space',","")
# df['Description'] = df['Description'].str.replace('"',"")
# df['Description'] = df['Description'].str.replace("' '","")
# df['Description'] = df['Description'].str.replace("'","")
    
```

```python
def set_default_feature_values(data):
    columns_list=["guests", "beds", "bathrooms", "bedrooms"]
    data[columns_list] = data[columns_list].fillna(1)
    return data

# Replace all listings with empty guests,beds,bathrooms or bedrooms as they should have a minimum of 1.
# df.loc[listingsDF['guests'].isnull(),['guests']] = 1 
# df.loc[listingsDF['beds'].isnull(),['beds']] = 1  
# df.loc[listingsDF['bathrooms'].isnull(),['bathrooms']] = 1  
# df.loc[listingsDF['bedrooms'].isnull(),['bedrooms']] = 1
# df['bathrooms'].loc[listingsDF['bathrooms'] < 1] = 1  
```

```python
def clean_tabular_data(data):
    data=remove_rows_with_missing_ratings(data)
    data=combine_description_strings(data)
    data=set_default_feature_values(data)

    return data
```
